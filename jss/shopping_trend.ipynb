### 나이 기준으로 연령대 구별하기

```python
df.loc[(df['Age'] >= 10) & (df['Age'] <= 19), 'Generation'] = 10
df.loc[(df['Age'] >=20) & (df['Age'] <30), 'Generation'] = 20
df.loc[(df['Age'] >=30) & (df['Age'] <40), 'Generation'] = 30
df.loc[(df['Age'] >=40) & (df['Age'] < 50), 'Generation'] = 40
df.loc[(df['Age'] >=50) & (df['Age'] < 60), 'Generation'] = 50
df.loc[(df['Age'] >=60) & (df['Age'] < 70), 'Generation'] = 60
df.loc[(df['Age'] >=70), 'Generation'] = 70
```

```python
df
```

```python
cus_total=df.groupby('Generation')['Purchase Amount (USD)'].sum().reset_index()
```

```python
cus_total
```

### 나이대 별로 총 얼마를 이용했는지 알려주는 그래프

```python
sns.barplot(data=cus_total,x='Generation',y='Purchase Amount (USD)')
plt.title('Generation Purchase')
plt.xlabel('Generation')
plt.ylabel('Purchase Amount (USD)')
```

# 그래프에서 볼수있는것
- 50대가 돈씀이 많다
- 20대 30대 40대 미세하지만 20대가 더 이용했다

```python
cus_total_gender=df.groupby(['Generation','Gender'])['Purchase Amount (USD)'].mean().reset_index()
```

```python
sns.barplot(data=cus_total_gender,x='Generation',y='Purchase Amount (USD)',hue='Gender')
plt.title('Gender Generation Purchase')
plt.xlabel('Gender Generation')
plt.ylabel('Purchase Amount (USD)')
```
