# 호텔 취소와 관련된 행동 분석

```python
def month_to_number(month_str):
    month_dict = {
        'January': 1,
        'February': 2,
        'March': 3,
        'April': 4,
        'May': 5,
        'June': 6,
        'July': 7,
        'August': 8,
        'September': 9,
        'October': 10,
        'November': 11,
        'December': 12
    }
    return month_dict[month_str]

df['arrival_date_month'] = df['arrival_date_month'].apply(month_to_number)
```

```python
df
```

```python
# 최종 예약 업데이트 일자와 도착일자 및 예약일자 간의 관계가 있는지 파악 위해 날자 형태 변경

from datetime import datetime
df['reservation_status_date'] = [datetime.strptime(date_str, '%d-%m-%y').strftime('%Y-%m-%d') for date_str in df['reservation_status_date']]
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
a = []
for i in df['arrival_date_month'].astype(str):
  if len(i)==1:
    i = '0'+i
  a.append(i)
df['arrival_date_month'] = a
```

```python
a = []
for i in df['arrival_date_day_of_month'].astype(str):
  if len(i)==1:
    i = '0'+i
  a.append(i)
```

```python
df['arrival_date_day_of_month'] = a
df['arrival_date'] = df['arrival_date_year'].astype(str)+df['arrival_date_month']+df['arrival_date_day_of_month']
df['arrival_date'] = pd.to_datetime(df['arrival_date'])
df['re-ar'] = (df['reservation_status_date'] -df['arrival_date']).dt.days.astype(int)
```

```python
# 도착일자와 업데이트 일자간의 차이 평균적으로 체크인 88일전 취소 발생

for i in df['reservation_status'].unique():
  print(i,df[df['reservation_status']==i]['re-ar'].mean())
```

```python
# 예약일이 평균적으로 5달전 이루워질 경우 취소가 많음

for i in df['reservation_status'].unique():
  print(i,df[df['reservation_status']==i]['lead_time'].mean())
```

```python
# reservation_status의 칼럼의 값이 Check-Out인 경우 is_canceled의 값은 0이므로 중복된 칼럼이라 생각 칼럼 삭제 취소나 노쇼의 경우 is_canceled은 1

print(df[df['reservation_status']=='Check-Out']['is_canceled'].value_counts())

df = df.drop(columns='is_canceled')
```

## 날짜에 따른 취소 비교

```python
dd = df[['arrival_date_year','arrival_date_week_number','stays_in_weekend_nights','stays_in_week_nights','reservation_status']]
dd
```

```python
# 년도 구분없이 arrival_date_week_number와 취소 관계분석

dd2 = dd[['arrival_date_week_number','reservation_status']].value_counts().reset_index().pivot_table(index='arrival_date_week_number',columns='reservation_status')
```

```python
# 주차마다 취소율 큰 차이를 보이고 있음, 주차별 취소 관계 존재 예상

dd2['rate'] = dd2[0]['Canceled'] / dd2[0]['Check-Out']
dd2.sort_values('rate',ascending=False)
```

```python
# 투숙 기간과 취소와의 관계

dd['week+weekend'] = dd['stays_in_weekend_nights'] + dd['stays_in_week_nights']
dd[['week+weekend','reservation_status']].value_counts().reset_index()
```

```python
dd['숙박기간'] = ''
dd.loc[dd['week+weekend']<=1,'숙박기간'] = '1박'
dd.loc[(dd['week+weekend']>1) & (dd['week+weekend']<=3),'숙박기간'] = '단기'
dd.loc[(dd['week+weekend']>3) & (dd['week+weekend']<=7),'숙박기간'] = '중기'
dd.loc[dd['week+weekend']>7,'숙박기간'] = '장기'
```

```python
# 머무는 기간이 1일 이하인 경우 취소율 제일 낮음

for i in dd['숙박기간'].unique():
  print(i, dd[dd['숙박기간']==i].reservation_status.value_counts(normalize=True).reset_index())
  print()
```

## 웨이팅, 고객 유형, 요금에 따른 취소와 관계 분석

```python
df2 = df[['reservation_status','days_in_waiting_list','customer_type','adr']]
```

```python
# 수치적으로 차이는 있지만 비율로 볼 때 Group로 올경우 취소 확률이 낮음

for i in df2['customer_type'].unique():
  print("고객 유형 : {}".format(i))
  print(df2[df2['customer_type']==i]['reservation_status'].value_counts(normalize=True))
  print()
```

```python
import numpy as np
np.quantile(df2['days_in_waiting_list'],0.9)
```

```python
# days_in_waiting_list의 경우 대다수가 0 이기 때문에 웨이팅 없음과 있음으로 구분 없음 :0 있음 : 1

print(df2.days_in_waiting_list.value_counts().reset_index(name='count'))

df2.loc[df2['days_in_waiting_list']>=1,'days_in_waiting_list'] = 1
```

```python
# 웨이팅 유무와 취소와의 관계는 큰 상관이 없는 것으로 보임

for i in df2['days_in_waiting_list'].unique():
  print("웨이팅 유무 : {}".format(i))
  print(df2[df2['days_in_waiting_list']==i].reservation_status.value_counts(normalize=True))
  print()
```

```python
# 숙박 요금의 경우 85 이하, 85초과 135이하, 135초과 3그룹으로 분류

print(np.quantile(df2['adr'],0.8))
print(np.quantile(df2['adr'],0.4))
```

```python
df2.loc[df2['adr']<=85,'adr'] = 0
df2.loc[(df2['adr'] <=135 ) & (df2['adr']>85),'adr'] =1
df2.loc[(df2['adr']>135),'adr'] = 2
```

```python
# 취소와 요금은 관계 없어 보임

for i in df2['adr'].unique():
  print("요금 유형 : {}".format(i))
  print(df2[df2['adr']==i].reservation_status.value_counts(normalize=True))
  print()
```
