## 1. 서울시 공공자전거 일별 대여건수 22년 1년치 데이터 불러오기

```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np


bike = pd.read_csv('/content/자전거.csv')
```

```python
bike.isnull().sum()
```

```python
bike.nunique()
```

```python
bike.rename(columns={
  "STA_LOC":"지역",
  "RENT_NM":"대여소명",
  "STAT_DATA":"대여일시",
  "RENT_CNT":"대여수",
  "RTN_CNT":"반납수"
},inplace=True)
```

```python
# nunique 값 1개인 칼럼 제거

bike = bike.iloc[:,:5]
```

```python
# 서울시 구별 대여소 개수
a = bike['지역'].value_counts().reset_index(name='count')
a.rename(columns={
    "index": "지역",
    "count":"대여소개수"},inplace=True)
```

```python
bike['대여수'] = bike['대여수'].astype(int)
bike['반납수'] = bike['반납수'].astype(int)
```

```python
# 대여건수 대비 반납건수
bike_group = bike.groupby('지역')['대여수','반납수'].sum()
bike_group = bike_group.reset_index()

bike_group['대여_반납_비율'] = bike_group['대여수'] / bike_group['반납수']
bike_group.sort_values('대여수',ascending=False)
```

```python
a.head()
```

```python
bike_merge = bike_group.merge(a,on='지역')
```

```python
# 대여소 개수 대비 대여수 비교
#

bike_merge['대여소개수_대여수'] =  bike_merge['대여수'] / bike_merge['대여소개수']
bike_merge.sort_values('대여소개수_대여수',ascending=False)
```

```python
bike['대여일시_time'] = pd.to_datetime(bike['대여일시'])
bike['월'] = bike['대여일시_time'].dt.month
```

```python
# 혹서기나 혹한기인 날이 대여수와 반납수가 적으며 날씨가 적당한 시기에 대여수 많음

bike.groupby(['월'])['대여수','반납수'].sum().reset_index().set_index('월').sort_values('대여수',ascending=False)
```

```python
df_com2 = df_com[['tm','지역','대여소명','대여수','반납수','month','maxTa','sumLrgEv','avgTa','ddMefs','maxPs','minTaHrmt','sumRn']]
```

```python
df_com2 = df_com2.rename(columns={
    'tm':'일자',
    'month':'월',
    'maxTa':'최고기온',
    'sumLrgEv':'합계대형증발량',
    'avgTa':'평균기온',
    'ddMefs':'일최심실적설',
    'maxPs':'최고해면기압',
    'minTaHrmt':'최저기온시각',
    'sumRn' : '일강수량'
})
```

```python
df_com2['요일'] = df_com2['일자'].dt.day_name()
df_com2['평일주말'] = ''
df_com2.loc[df_com2['요일'].isin(['Saturday', 'Sunday']),'평일주말'] = '주말'
df_com2.loc[~df_com2['요일'].isin(['Saturday', 'Sunday']),'평일주말'] = '평일'
```

```python
# 전반적인 요일별 대여수를 봤을 때 대여수가 평일이 더 많음
# 이는 서울시 공공자전거는 일상적인 루틴이 많은 출퇴근, 등하교 등 많이 이용됨을 예상

df_com2.groupby('요일')['대여수'].sum().reset_index().sort_values('대여수',ascending=False)
```

```python
end =df_com2[df_com2['평일주말']=='주말'].groupby(['월','평일주말'])['대여수'].sum().reset_index()
day = df_com2[df_com2['평일주말']=='평일'].groupby(['월','평일주말'])['대여수'].sum().reset_index()
```

```python
# 4월과 9월이 다른 월에 비해 평일 대비 주말 대여 비율이 높음

dayend = day.merge(end,on='월')
dayend['비율'] = dayend['대여수_y'] / dayend['대여수_x']
dayend.sort_values('비율',ascending=False)
```

```python
april = df_com2[df_com2['월']==4].groupby('일자').mean().reset_index()
april['요일'] = april['일자'].dt.day_name()
april['평일주말'] = ''
april.loc[april['요일'].isin(['Saturday', 'Sunday']),'평일주말'] = '주말'
april.loc[~april['요일'].isin(['Saturday', 'Sunday']),'평일주말'] = '평일'
```

```python
april['최저기온시각'] = april['최저기온시각'].astype(int)
april['최저기온시간'] = [m / 60 % 24 for m in april['최저기온시각']]
april['대여수'] = april['대여수'].astype(int)
april['반납수'] = april['반납수'].astype(int)
```

```python
# 전반적으로 최저기온시각, 일강수량, 이 높을수록 대여수가 적으며 최고기온, 합계대형증발량이 높을 수록 대여수 높음

april.sort_values('대여수',ascending=False)
```

```python
# 전체 월을 비교했을 때와 특정기간에서의 상관관계는 크게 다르게 나옴
# 양의 상관관계를 보였던 부분들이 여름에는 음의상관관계

df_com2[df_com2['월'].isin([6,7,8])].groupby('월').mean().corr()
```

```python
# 여름과 겨울에서 기온의 상관관계가 정 반대로 보임


df_com2[df_com2['월'].isin([12,1,2])].groupby('월').mean().corr()
```

```python
# 여름에는 한강공원, 여의나루, 올림픽공원 같은 곳이 높으나 겨울에는 적음

df_com2[df_com2['월'].isin([12,1,2])].groupby('대여소명')['대여수','반납수'].sum().sort_values('대여수',ascending=False).head(20)
```

```python
rental = df_com2.groupby(['대여소명'])['대여수'].sum().reset_index()
retur = df_com2.groupby(['대여소명'])['반납수'].sum().reset_index()
```

```python
# 대여소 기준 대여보다 반납이 많은 곳을 보면 대부분 지하철역 근처
# 지하철 이용시 서울 공공자전거 많이 사용 하는것을 유추

re_merge =  rental.merge(retur,on=['대여소명'])
re_merge['반납-대여'] = re_merge['반납수'] - re_merge['대여수']
re_merge.sort_values('반납-대여',ascending=False).head(40)
```

```python
#  대여수가 더 많은 경우는 대부분 거주지로 추정 거주지에서 대여 후 지하철역 에서 반납 예상

re_merge.sort_values('반납-대여').head(40)
```

```python
c = df_com2.groupby(['대여소명','평일주말'])['대여수'].sum().reset_index()
week = c[c['평일주말']=='평일'].merge(c[c['평일주말']=='주말'],on='대여소명')
week['주말_평일_대여'] = week['대여수_y'] - week['대여수_x']
```

```python
# 주말에 대여를 많이하는 곳은 공원과 같이 주말에 나들이 많이 하는 곳이 많음

week[week['대여수_y']>=1000].sort_values('주말_평일_대여',ascending=False).head(40)
```

### 1.자전거 대여 행태
#####  └ 대부분의 대여소에서 대여수와 반납수가 비슷
#####  └ 대여가 반납보다 많은 경우는 아파트 등 주거단지 근처
#####  └ 반대로 반납이 대여보다 많은 경우는 지하철과 같은 교통 수단 근처
#####  └ 주말과 평일 비교시 평일에는 잘 보이지 않던 공원 같이 자전거 타기 좋은 곳이 높음

### 2. 자전거 대여와 날씨와의 관계
#####   └ 여름과 겨울로 비교 진행
#####   └ 여름 양의 상관관계 : 합계대형증발량, 최고해면기압
#####   └ 음의 상관관계 : 기온, 최저기온시각, 일강수량
#####   └ 겨울 양의 상관관계 : 기온, 합계대형증발량, 최고해면기압
#####   └ 겨울 음의 상관관계 : 일최심실적설, 일강수량
#####   └ 여름과 겨울 둘다 합계대형증발량, 최고해면기압 양의상관관계 가지고있으며 기온은 반대
