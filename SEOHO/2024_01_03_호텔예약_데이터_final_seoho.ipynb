# 호텔 취소와 관련된 행동 분석

```python
def month_to_number(month_str):
    month_dict = {
        'January': 1,
        'February': 2,
        'March': 3,
        'April': 4,
        'May': 5,
        'June': 6,
        'July': 7,
        'August': 8,
        'September': 9,
        'October': 10,
        'November': 11,
        'December': 12
    }
    return month_dict[month_str]

df['arrival_date_month'] = df['arrival_date_month'].apply(month_to_number)
```

```python
# 최종 예약 업데이트 일자와 도착일자 및 예약일자 간의 관계가 있는지 파악 위해 날자 형태 변경

from datetime import datetime
df['reservation_status_date'] = [datetime.strptime(date_str, '%d-%m-%y').strftime('%Y-%m-%d') for date_str in df['reservation_status_date']]
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

### 다음 주 할일

취소와 관련이 있다고 파악된 칼럼을 통해 모델링 진행

- country : 국가
- re-ar : 도착일자 - 업데이트 일자
- lead_time : 예약일
- company : 예약에 참여한 회사

- stays_in_weekend_nights : 주말 숙박기간
- stays_in_week_nights : 주중 숙박기간
- customer_type : 고객 유형
- total_of_special_requests : 요청 수
- required_car_parking_spaces : 주차 요청 수
- reservation_status : 취소 유무

```python
df.loc[df['reservation_status']!='Check-Out','reservation_status'] = 1
df.loc[df['reservation_status']=='Check-Out','reservation_status'] = 0
```

```python
X = df[['re-ar','lead_time','company','arrival_date_week_number','stays_in_weekend_nights','stays_in_week_nights','customer_type','total_of_special_requests','required_car_parking_spaces']]
Y = df[['reservation_status']]
```

```python
# company 가 있는경우 1 없는경우 0

X.loc[X['company'].isnull(),'company'] = 0
X.loc[X['company']!=0,'company'] = 1
```

```python
df['customer_type'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

encoder_customer_type = LabelEncoder()
X['customer_type'] = encoder_customer_type.fit_transform(X['customer_type'])
onehot_encoder = OneHotEncoder(sparse=False, drop='first')
encoded_features = onehot_encoder.fit_transform(X[['customer_type']])
ds = pd.DataFrame(encoded_features,columns=['cate_0','cate_1','cate_2'])

X = pd.concat([X,ds],axis=1)
X = X.drop(columns='customer_type')
```

```python
Y = Y.astype(int)
```

```python
from sklearn.model_selection import train_test_split
Train_X, Test_X, Train_Y, Test_Y = train_test_split(X, Y)
```

```python
from sklearn.model_selection import ParameterGrid
from sklearn.metrics import f1_score

# 모델 정의
models = [
    LogisticRegression,
    KNeighborsClassifier,
    SVC,
    DecisionTreeClassifier,
    RandomForestClassifier,
    GradientBoostingClassifier,
    GaussianNB,
    MLPClassifier
]

# 모델별 파라미터 그리드 생성
param_grid = {
    LogisticRegression: ParameterGrid({"C": [0.1, 1, 10], "max_iter": [100, 200, 300]}),
    KNeighborsClassifier: ParameterGrid({"n_neighbors": [3, 5, 7], "metric": ['euclidean', 'manhattan']}),
    SVC: ParameterGrid({"C": [0.1, 1, 10], "kernel": ['rbf', 'linear']}),
    DecisionTreeClassifier: ParameterGrid({"max_depth": [None, 10, 20], "min_samples_split": [2, 5, 10]}),
    RandomForestClassifier: ParameterGrid({"n_estimators": [50, 100, 200], "max_depth": [None, 10, 20]}),
    GradientBoostingClassifier: ParameterGrid({"n_estimators": [50, 100, 200], "learning_rate": [0.01, 0.1, 0.2]}),
    GaussianNB: ParameterGrid({}),
    MLPClassifier: ParameterGrid({"hidden_layer_sizes": [(10,), (20,), (30,)], "alpha": [0.0001, 0.001, 0.01]})
}

def tune_hyperparameters(model_func, param_grid, Train_X, Train_Y, Test_X, Test_Y):
    best_model = None
    best_param = None
    best_score = -1

    for param in param_grid:
        model = model_func(**param).fit(Train_X, Train_Y)
        pred_Y = model.predict(Test_X)
        score = f1_score(Test_Y, pred_Y, average='micro')

        if score > best_score:
            best_model = model
            best_param = param
            best_score = score

    return best_model, best_param, best_score



# 각 모델에 대한 하이퍼파라미터 튜닝 수행
best_models = {}
for model_func in models:
    grid = param_grid.get(model_func, ParameterGrid({}))
    best_model, best_param, best_score = tune_hyperparameters(model_func, grid, Train_X, Train_Y, Test_X, Test_Y)
    best_models[model_func] = {'best_model': best_model, 'best_param': best_param, 'best_score': best_score}


for model_func, result in best_models.items():
    print(f"Model: {model_func.__name__}")
    print(f"Best Parameters: {result['best_param']}")
    print(f"Best F1 Score: {result['best_score']}\n")
```
